```python
#Adri Jackson, with assistance from Dr. Jonathan Niefhof, UNH
import datetime
import numpy
import matplotlib.pyplot as plt

def read_data(filename):
    recarray = numpy.core.records.fromfile(
        filename,
        formats="i2,i2,i2,u2,i2,i2,i2,i2,i2,i2,i2,i2,i2,i2,i2,(2580,)i2,i2,(44,)i2", 
        names = "day,hour,minute,second,year,geodetic_lat,geographic_long,alt,geolat_110,geolong_110,corrected_mlat,"
                "corrected_mlong,mlt_hour,mlt_minute,mlt_second,channels,indicator,zero_fill",
        byteorder='>')
    channels_dtype = [
        ('h', '>i2'),
        ('m', '>i2'),
        ('s', '>u2'),
        ('e', '>i2', (20,)),
        ('i', '>i2', (20,)),
    ]
    # Reinterpret the channels, and make 1D in time
    channels = numpy.ascontiguousarray(recarray["channels"])
    channels = channels.view(channels_dtype).reshape((-1,))
    unfolded_dtype = [
        ('day', '>i2'),
        ('hour', '>i2'),
        ('minute', '>i2'),
        ('second', '>u2'),
        ('year', '>i2'),
        ('geodetic_lat', '>i2'),
        ('geographic_long', '>i2'),
        ('alt', '>i2'),
        ('geolat_110', '>i2'),
        ('geolong_110', '>i2'),
        ('corrected_mlat', '>i2'),
        ('corrected_mlong', '>i2'),
        ('mlt_hour', '>i2'),
        ('mlt_minute', '>i2'),
        ('mlt_second', '>i2'),
    ] \
    + [
        ('h', '>i2'),
        ('m', '>i2'),
        ('s', '>f'),  # has to be float if indicator
        ('e', '>i2', (20,)),
        ('i', '>i2', (20,)),
    ] \
    + [
        ('ssd', 'f4'),  # seconds start of day
        ('dt', 'O'),  # datetime
        ('mlt', 'f4'),  # unfolded MLT
        ('minute_start', bool),  # this was start of a one-minute record
        ('mlat_deg', 'f4'),  # magnetic latitude in degrees
        ('mlon_deg', 'f4'),  # magnetic longitude in degrees
        ('geolat_110_deg', 'f4') # Geographic latitude at 110 km altitude and on the same magnetic field line as the DMSP spacecraft, in degrees
        
    ]
    # By-second array
    data = numpy.empty((60 * len(recarray),), dtype=unfolded_dtype)
    # Fields that are the same between original recarray and final unfolded array
    same_fields = [k for k in recarray.dtype.fields if k in data.dtype.fields]
    # Copy over the data that are the same
    data[same_fields] = numpy.repeat(recarray[same_fields], 60)
    # Same trick for the channels array
    same_fields = [k for k in channels.dtype.fields if k in data.dtype.fields]
    data[same_fields] = channels[same_fields]
    # Convert ms to s
    idx = numpy.repeat(recarray['indicator'].astype(bool), 60)
    data['s'][idx] /= 1000
    # Note start-of-record
    data['minute_start'] = False
    data['minute_start'][::60] = True
    # Remove all-zero per-second records, which are fill at the end of each minute record
    idx = (recarray['channels'].reshape(-1, 43) != 0).any(axis=-1)
    data = data[idx]
    # And make the derived fields
    # Leaving both types of hours/minutes/seconds in the output, can always change that
    data['ssd'] = (data['h'] * 60. + data['m']) * 60 + data['s']
    yyyy = 2000  
    data['dt'] = numpy.frompyfunc(lambda d, ssd: datetime.datetime(yyyy, 1, 1) + datetime.timedelta(days=d, seconds=ssd),
                                 nin=2, nout=1)(data['day'], data['ssd'])
    # doing this order gets the casting right
    data['mlt'] = (data['mlt_second'].astype('f4') / 60 + data['mlt_minute']) / 60 + data['mlt_hour']
    mlat = data['corrected_mlat']
    data['mlat_deg'] = (mlat.astype('f4') - numpy.choose(mlat > 1800, (900., 4995.)))/ 10.
    geo110_deg = data['geolat_110']
    data['geolat_110_deg'] = (geo110_deg.astype('f4') - numpy.choose(geo110_deg > 1800, (900., 4995.)))/ 10.
    return data

#filter data for when when the satellite is between 72 and 82 mlat and extract electron data, time, magnetic latitude, 
#and magnetic local time (hour and minute) 
def filter_data(data):
    filtered_dat = numpy.empty((1,20))
    filtered_times = []
    mlat = numpy.array(1)
    mlt_hr = numpy.array(1)
    mlt_mi = numpy.array(1)
    for i in range(len(data)):
        if data[i][24] >= 72 and data[i][24] <= 82: 
            filtered_dat = numpy.vstack((filtered_dat,data[i][18]))
            filtered_times.append(str(datetime.time(data[i][15],data[i][16],data[i][17])))
            mlat = numpy.append(mlat,data[i][24])
            mlt_hr = numpy.append(mlt_hr,data[i][12])
            mlt_mi = numpy.append(mlt_mi,data[i][13])
    filt_dat = numpy.delete(filtered_dat,0,0)
    mlat = numpy.delete(mlat,0)
    mlt_hr = numpy.delete(mlt_hr,0)
    mlt_mi = numpy.delete(mlt_mi,0)
    return filt_dat, filtered_times, mlat, mlt_hr, mlt_mi

#convert electron raw data to "counts" as defined in the documentation
def calc_counts(a):
    X = a%32
    Y = (a-X)/32.0
    counts = ((X + 32)*(2**Y))-33
    return counts

#create array of energy bins, 949 is repeated as part of the documentation
E = [30,44,65,95,139,204,300,440,646,949,949,1392,2040,3000,4400,6460,9450,13900,20400,30000]
#space between energy channels (in eV, used for calculation)
DE_i = [14,17.5,25.5,37,54.5,80.5,118,173,254.5,373,373,545.5,804,1180,1730,2525,3720,5475,8050,9600]
#accumulation time for detector
Dt = 0.098
#geometric factor for each energy bin of detector
G_i = [0.0002496,0.0006463,0.001359,0.002598,0.004827,0.008563,0.01232,0.01777,0.02615,0.03644,0.0572,0.0721,0.0908,0.114,0.144,0.181,0.228,0.287,0.362,0.456]

#calculations of differential number flux, differential energy flux, integrated number flux, and integrated energy flux
#as defined in the documentation. The last quantity calculated is the average differential number flux. This is to get a 
#shape of the average flux per energy bin through plotting. Each set of values is stored in an appropriately shaped array
def calc_sci_units(a):
    J_i = numpy.zeros((len(a),20))
    for i in range(len(a)):
        for j in range(len(G_i)):
            J_i[i,j] += a[i,j]/(G_i[j]*Dt)
        
    JE_i = numpy.zeros((len(a),20))
    for i in range(len(a)):
        for j in range(len(E)):
            JE_i[i,j] += J_i[i,j]*E[j]
        
    J = numpy.zeros(len(a))
    k = 0
    for i in range(len(a)):
        for j in range(len(DE_i)):
            k += J_i[i,j]*DE_i[j]
        J[i] += k
        k = 0
    
    JE = numpy.zeros(len(a))
    k = 0
    for i in range(len(a)):
        for j in range(len(DE_i)):
            k += JE_i[i,j]*DE_i[j]
        JE[i] += k
        k = 0
    
    J_i_avg = numpy.mean(J_i, 0)
    return J_i, JE_i, J, JE, J_i_avg

#this functions takes the hour and minute components of magnetic local time, combines them as a sum of minutes, then
#interpolates values between each minute, since only per-minute values were given. After interpolation, the 
#minute/hour components are separated again and formatted into an hh:mm time
def mlt_interp(mlt_hr,mlt_mi):
    mlt_sum = numpy.empty(len(mlt_hr))
    for i in range(len(mlt_hr)):
        mlt_sum[i] = mlt_hr[i]*60 + mlt_mi[i]
    
    x_k = numpy.array(1)
    x_val = numpy.array(1)
    x_u = numpy.array(1)

    for i in range(len(mlt_sum)):
        if i%60 == 0:
            x_k = numpy.append(x_k, i)
            x_val = numpy.append(x_val, mlt_sum[i])
            x_u = numpy.append(x_u, i)
        else:
            x_u = numpy.append(x_u, i)
        
    x_k = numpy.delete(x_k,0)
    x_val = numpy.delete(x_val,0)
    x_u = numpy.delete(x_u,0)

    x_interp = numpy.interp(x_u,x_k,x_val)
    x_interp_hr = x_interp//60
    x_interp_min = numpy.rint(x_interp%60)
    mlt_recom = []
    for i in range(len(mlt_sum)):
        x = '{0:02.0f}{1}{2:02.0f}'.format(int(x_interp_hr[i]),':',int(x_interp_min[i]),sep='')
        mlt_recom.append(x)
    return mlt_recom

#this functions takes the time, mlat, and mlt to format into a list of 3-line strings for labelling plots 
def create_labels(time,mlat,mlt):
    labels = []
    for i in range(len(time)):
        labels.append('{0}{1}{2:0.2f}{3}{4}'.format(time[i],'\n',mlat[i],'\n',mlt[i]))
    return labels

x_ax_label = '{0}{1}{2}{3}{4}'.format('UTC','\n','Mlat','\n','MLT')

#This function is an alternative way to calculate the integrated number flux ("J" in the previous cell) but first removing
#a number of energy levels (columns), s, to simulate the ERPA screening out these electrons. The function takes the differential
#flux array, the energy bin spacing array, and the integer for how many energy bins to remove, as arguments. After removing the
#correct number of bins, the modified integrated flux is calculated, and then returned 
def screen_energy(diff, DE, s):
    diff = numpy.delete(diff,numpy.s_[0:s:1],1)
    DE = numpy.delete(DE,numpy.s_[0:s:1])
    Int = numpy.zeros(len(diff))
    k = 0
    for i in range(len(diff)):
        for j in range(len(DE)):
            k += diff[i,j]*DE[j]
        Int[i] += k
        k = 0
    return Int

#this block allows the user to enter 1 or more data files and calculate the average or maximum electron count value.
#It will repeat if the user chooses to calculate something else at the end of the loop
choice = "y" #calculate value at least once   
while choice == "y" or choice == "Y":
    try: #try/except block for error handling
        #prompt the user to enter the file name
        data = read_data(input("Please enter the name of the data file you are working with: "))
        #prompt the user to ask if they would like to add anaother file 
        another_file = input("Would you like to add another file? y/n: ")
        # if the user wants to add another file this block excecutes
        while another_file == "y" or another_file == "Y":
            try: # nested try/except block for error handling
                # reads the new data file and processes it into an array
                next_file = read_data(input("Please enter the name of the data file: "))
                # concatenates (combines) the arrays into one array
                data = numpy.hstack([data, next_file])
                another_file = input("Would you like to add another file? y/n: ")
            except:
                # catches errors in user inputs
                print("Error: incorrect file name")
                # gives the user an option to exit the loop
                another_file = input("Are you sure you would like to add another file? y/n: ")
    except: 
        print("Error: please enter the correct file name")
    #inform the user the data is being processed
    print("Processing data. This will take longer the more files that were added.")
    #call necessary functions to perform calculation
    e_array, time, mlat, mtl_hr, mlt_mi = filter_data(data)
    counts = numpy.apply_along_axis(calc_counts,0, e_array)
    e_counts = (counts[:,[16,17,18,19,12,13,14,15,8,9,10,11,4,5,6,7,0,1,2,3]])
    #mlt = mlt_interp(mlt_hr,mlt_mi)
    #x_labels = create_labels(time,mlat,mlt)
    J_i, JE_i, J, JE, J_i_avg = calc_sci_units(e_counts)
    #loop to catch invalid calculation choice
    calc = '0'
    while calc != '1' and calc != '2':
        #the prompt the user to ask if they would like to calculate the max or mean
        calc = input("Would you like to find the maximum (1) or average (2) electron flux? Please enter 1 or 2: ")
        if calc == '1':
            #calculate max and display it if chosen
            max_val = numpy.max(J)
            print('{0}{1:.2E}{2}'.format('The maximum value for the range of dates you input is: ',max_val,' electron counts'))
        elif calc == '2':
            #calculate mean and display it if chosen
            mean_val = numpy.mean(J)
            print('{0}{1:.2E}{2}'.format('The average value for the range of dates you input is: ',mean_val,' electron counts'))
        else:
            #if the user enters an invalid choice, inform them, and restart the loop
            print("Error: please enter 1 or 2.")
    #after displaying calculated value, prompt the user to see if they would like to perform a new calculation (repeat loop)        
    choice = (input("Would you like to caclulate anything else? y/n: "))
            
```

```python
#plot April data without filtering. x-axis labels are limited for legibility and compilation time. The max is printed for each
#day. All 30 days in April are plotted
data = read_data('Apr_01')
e_array, time, mlat, mlt_hr, mlt_mi = filter_data(data)
counts = numpy.apply_along_axis(calc_counts,0, e_array)
e_counts = (counts[:,[16,17,18,19,12,13,14,15,8,9,10,11,4,5,6,7,0,1,2,3]])
mlt = mlt_interp(mlt_hr,mlt_mi)
x_labels = create_labels(time,mlat,mlt)
J_i, JE_i, J1, JE, J_i_avg = calc_sci_units(e_counts)

fig, axs = plt.subplots(1, 1, figsize=(25,5))

axs.scatter(x_labels, J1)
axs.xaxis.set_tick_params(labelsize=10)
axs.xaxis.set_major_locator(plt.MaxNLocator(25))
axs.set_xlim(0,len(time))
axs.set_ylim(0,1.1*10**10)
print(numpy.max(J1))
```
